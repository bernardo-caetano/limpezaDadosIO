```python
import pandas as pd 
df=pd.read_csv('gdrive/My Drive/Trabalho/Estágio LCA/Trabalho LCA/Dados/planta_full_time(titulo certo).csv', sep = ';', encoding = 'utf-8')
df.head()
```

```python
df_no_time=df.drop(columns=['Timestamp(ms)'])
df_no_time.tail()
```

**código com planilha final**

```python
df_final=df_no_time.loc[:0]
i=0
j=0
for i in range((len(df_no_time.index))-1):
  aux=np.array(df_no_time.loc[i:i])
  aux_prox=np.array(df_no_time.loc[(i+1):(i+1)])
  for j in range(len(aux[0])):
    if aux[0][j] != aux_prox[0][j]:
      df_final=df_final.append(df_no_time.loc[(i+1):(i+1)], ignore_index=True)
      break
print(len(df_final.index))
#df_teste.head()
df_final.head()
```

**Fazendo download da planilha final em formato excel**

```python
df_final.to_excel('transicoes_io.xlsx')
files.download('transicoes_io.xlsx')
```

**Fazendo download da planilha final em formato csv**

```python
df_final.to_csv('transicoes_io.csv',index=False)
files.download('transicoes_io.csv')
```
